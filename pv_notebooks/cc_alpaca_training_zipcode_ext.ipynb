# Extension Zipcode
### This Notebook is used to add the extension zipcode column to the synthetic data created. This addition is not Differential Privacy perturbed code, rather just another synthetic extension to the original data, which did not have this code

```python
with open('../datasets/fraudTrain.csv') as fd:
    numcols = len(fd.readline().split(','))
df = pd.read_csv('../datasets/fraudTrain.csv', usecols=range(1,numcols))
```

```python
display(df.head(5))
```

```python
# Add a new zip_ext column filled with np.nan
indx = df.columns.get_loc('zip')
indx += 1
df.insert(loc=indx, column='zip_ext', value=0)
df['zip_ext'] = df['zip_ext'].astype(int)
display(df.head(5))
```

```python
# Construct unique identities 
identities = df.groupby(['first', 'last']).size().reset_index()
# Iterate over each identity and find matching columns
# and assing a new zipcode extension
for r in identities.iterrows():
    df.loc[(df['first'] == r[1]['first']) & (df['last'] == r[1]['last']), 'zip_ext'] =  np.random.randint(1001, 10000)
```

```python
df.to_csv('../datasets/fraudTrainWithExtZip.csv')
```
